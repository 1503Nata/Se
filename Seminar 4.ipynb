### 1.1. Скачать следующие данные: kc-house-data и laptop_price. ###

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('kc_house_data.csv', encoding = '1251')
df
```

### 1.2. Изучите стоимости недвижимости. ###

```python
plt.figure(figsize=(7, 10))
plt.hist(df['price'])
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость объектов')
plt.ylabel('Количество объектов');
```

### 1.3. Изучите распределение квадратуры жилой. ###

```python
plt.figure(figsize=(7, 10))
sns.histplot(df['sqft_living'], bins=20)
plt.title('Распределение жилой площади объектов')
plt.xlabel('Жилая площадь объекта, м2')
plt.ylabel('Количество объектов');
```

### 1.4. Изучите распределение года постройки. ###

```python
plt.figure(figsize=(7, 10))
sns.histplot(df['yr_built'], bins=30)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество объектов');
```

## Задание № 2 ##

### 2.1. Изучите распределение домов от наличия вида на набережную. ###
#### Постройте график. ####
#### Сделайте выводы ####

```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(5, 7))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['Объекты не имеют вида на набережную', 'Объекты имеют вид на набережную']);
```

#### Вывод: 99.2 % объекта недвижимости не имеют сида на набережную, что оказывает существенное влияние на их стоимость. ####

###  2.2. Изучите распределение этажей домо, ###

```python
data = df['floors'].value_counts()

plt.figure(figsize=(5, 10))
plt.pie(data, autopct='%1.1f%%', labels=data.index)
plt.title('Распределение объектов с разной этажностью');
```

#### Вывод: Наибольшее количество проданых объектов имееют 1 этаж - 49,4% от всех заключенных сделок, что говорить о повышенном спросе. На втором месте (38,1%) стоят дома с 2-мя этажами. Небольшое количество сделок с объектами, имеющими подвальные помещения говорят о небольшом спросе на такой вид объектов, так как необоснованно повышает стоимость. ####

### 2.3. Изучите распределение состояния домов. ###


```python
data = df['condition'].value_counts()

plt.figure(figsize=(5, 10))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

#### Вывод: Большенство объектов имеет отметку состояния "3", что говорит о их удовлетворительном состоянии на момент оценки. ####

## Задание № 3. ##

### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями. ###


```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('Стоимость объекта')
plt.ylabel('Оценка объекта')
plt.title('Зависимость стоимости объекта оценки конструкций и дизайна');
```

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('Стоимость объекта')
plt.ylabel('Этажность объекта')
plt.title('Зависимость стоимости объекта от этажности');
```

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Стоимость объекта')
plt.ylabel('Оценка вида')
plt.title('Зависимость стоимости объекта от оценки вида');
```

#### Выводы: Анализ заключенных сделок с объектами недвижимости, позволяют сделать вывод, что достаточно много факторов оказывают существенное влияние на стоимость. Однако некоторые из них могут неоправданно повышать стоимость объекта, что приводит к снижению спроса. ####
